```python
import pandas as pd
```

```python
pred = pd.read_csv("epic100_validation_predictions_final.csv")  # prediction file
val = pd.read_csv("EPIC_100_validation_label.csv")              # ground truth
```

```python
merged = pd.merge(pred, val, on="narration_id", how="inner", suffixes=('_pred', '_true'))
merged.rename(columns={
    "participant_id_true": "participant_id",
    "video_id_true": "video_id",
    "start_frame_true": "start_frame",
    "stop_frame_true": "stop_frame",
    "noun_class_true": "noun_class",
    "verb_class_true": "verb_class"
}, inplace=True)
```

```python
merged["noun_correct"] = merged["pred_noun_class"] == merged["noun_class"]
merged["verb_correct"] = merged["pred_verb_class"] == merged["verb_class"]
merged["action_correct"] = merged["noun_correct"] & merged["verb_correct"]
```

```python
total = len(merged)
print(f"Total examples: {total}")
print(f"Noun accuracy: {merged['noun_correct'].mean():.4f}")
print(f"Verb accuracy: {merged['verb_correct'].mean():.4f}")
print(f"Action accuracy (noun+verb): {merged['action_correct'].mean():.4f}\n")
```

```python
accuracy_per_action_combo = merged.groupby(["noun", "verb"])["action_correct"].mean().reset_index()
accuracy_per_action_combo.rename(columns={"action_correct": "accuracy"}, inplace=True)

perfect_combos = accuracy_per_action_combo[accuracy_per_action_combo["accuracy"] == 1]
zero_combos = accuracy_per_action_combo[accuracy_per_action_combo["accuracy"] == 0]

print("Perfect noun+verb combinations (accuracy=1):")
print(perfect_combos.to_string(index=False), "\n")
print("Completely wrong noun+verb combinations (accuracy=0):")
print(zero_combos.to_string(index=False), "\n")
```

```python
accuracy_per_noun = merged.groupby("noun")["noun_correct"].mean().reset_index()
accuracy_per_noun.rename(columns={"noun_correct": "accuracy"}, inplace=True)

perfect_nouns = accuracy_per_noun[accuracy_per_noun["accuracy"] == 1]
zero_nouns = accuracy_per_noun[accuracy_per_noun["accuracy"] == 0]

print("Perfect nouns (accuracy=1):")
print(perfect_nouns.to_string(index=False), "\n")
print("Completely wrong nouns (accuracy=0):")
print(zero_nouns.to_string(index=False), "\n")
```

```python
accuracy_per_verb = merged.groupby("verb")["verb_correct"].mean().reset_index()
accuracy_per_verb.rename(columns={"verb_correct": "accuracy"}, inplace=True)

perfect_verbs = accuracy_per_verb[accuracy_per_verb["accuracy"] == 1]
zero_verbs = accuracy_per_verb[accuracy_per_verb["accuracy"] == 0]

print("Perfect verbs (accuracy=1):")
print(perfect_verbs.to_string(index=False), "\n")
print("Completely wrong verbs (accuracy=0):")
print(zero_verbs.to_string(index=False), "\n")
```
